```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("StudentsPerformance.csv")
data.head()
```

```python
data.info()
data.describe()
data.isnull().sum()
```

```python
data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]
data.head()
```

```python
sns.countplot(x='gender', data=data)
plt.title('Gender Distribution')
plt.show()
```

```python
data.groupby('gender')[['math_score','reading_score','writing_score']].mean()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='parental_level_of_education', y='math_score', data=data)
plt.title('Parental Education vs Math Score')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='test_preparation_course', y='reading_score', data=data)
plt.title('Test Preparation Course vs Reading Score')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data[['math_score','reading_score','writing_score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Scores')
plt.show()
```
